# Inferential Statistics - Chi-Square Test

## Business Question
Is there a statistically significant relationship between Overtime and Attrition?

## Hypotheses

H₀: Overtime and Attrition are independent.

H₁: Overtime and Attrition are associated.

## Statistics Used

- Contingency Table
- Chi-Square Test of Independence
- P-value
- Degrees of Freedom

## Business Importance

Determines whether the relationship between overtime and employee attrition is statistically significant rather than due to random variation.

```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(
    df["OverTime"],
    df["Attrition"]
)

contingency_table
```

```python
chi2, p, dof, expected = chi2_contingency(contingency_table)
```

```python
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
```

There is statistically significant evidence that OverTime and Attrition are associated.

```python
alpha = 0.05

if p < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

# Inferential Statistics - Independent t-Test

## Business Question
Is the average monthly income significantly different between employees who stayed and those who left?

## Hypotheses

H₀: The average monthly income is the same for both groups.

H₁: The average monthly income is different between the two groups.

## Statistics Used

- Independent Samples t-Test (Welch's t-test)
- t-Statistic
- P-value

## Business Importance

Determines whether salary differences between employees who stayed and those who left are statistically significant.

```python
from scipy.stats import ttest_ind

# Create two groups
salary_yes = df[df["Attrition"] == "Yes"]["MonthlyIncome"]
salary_no = df[df["Attrition"] == "No"]["MonthlyIncome"]

# Perform Welch's t-test
t_stat, p_value = ttest_ind(
    salary_yes,
    salary_no,
    equal_var=False
)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Decision
alpha = 0.05

if p_value < alpha:
    print("\nConclusion: Reject the Null Hypothesis")
    print("The average monthly income differs significantly between employees who stayed and those who left.")
else:
    print("\nConclusion: Fail to Reject the Null Hypothesis")
    print("No statistically significant difference in average monthly income was found.")
```

# Inferential Statistics - One-Way ANOVA

## Business Question
Is the average monthly income different across departments?

## Hypotheses

H₀: All departments have the same average monthly income.

H₁: At least one department has a different average monthly income.

## Statistics Used

- One-Way ANOVA
- F Statistic
- P-value

## Business Importance

Determines whether department is associated with differences in employee salaries.

```python
from scipy.stats import f_oneway
```

```python
sales = df[df["Department"]=="Sales"]["MonthlyIncome"]

rnd = df[df["Department"]=="Research & Development"]["MonthlyIncome"]

hr = df[df["Department"]=="Human Resources"]["MonthlyIncome"]
```

```python
f_stat, p_value = f_oneway(
    sales,
    rnd,
    hr
)
```

```python
print("F Statistic:", f_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

# Inferential Statistics - Confidence Interval

## Business Question
What is the likely range for the true average monthly income?

## Statistics Used

- Sample Mean
- Standard Error
- t Distribution
- 95% Confidence Interval

## Business Importance

Provides a range of plausible values for the population mean, helping decision-makers understand the precision of the estimate.

```python
std = df["MonthlyIncome"].std()

n = len(df)
```

```python
standard_error = std / np.sqrt(n)
```

```python
confidence = 0.95

ci = t.interval(
    confidence,
    df=n-1,
    loc=mean_salary,
    scale=standard_error
)

print(ci)
```

We estimate the average monthly income to be between ₹6262 and ₹6743 with 95% confidence.

```python
print("Mean Salary:", mean_salary)
print("95% Confidence Interval:", ci)
```

# Machine Learning - Train Test Split

## Objective

Split the dataset into training and testing sets.

## Why?

The model should learn from one dataset and be evaluated on unseen data.

## Parameters

- test_size = 0.20
- random_state = 42

## Business Importance

Ensures the model generalizes well to new employees.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
# Create model
model = LogisticRegression(max_iter=1000)
# Train model
model.fit(X_train, y_train)
# Predict classes
y_pred = model.predict(X_test)
# Predict probabilities
y_prob = model.predict_proba(X_test)
print("First 5 Predictions:")
print(y_pred[:5])
print("\nFirst 5 Probabilities:")
print(y_prob[:5])
```

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve
)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
print(classification_report(y_test, y_pred))
```

```python
y_prob = model.predict_proba(X_test)[:,1]

auc = roc_auc_score(y_test, y_prob)

print(auc)
```

```python
fpr, tpr, _ = roc_curve(y_test, y_prob)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr)
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)

dt.fit(X_train,y_train)

dt_pred = dt.predict(X_test)

print(classification_report(y_test,dt_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)

rf.fit(X_train,y_train)

rf_pred = rf.predict(X_test)

print(classification_report(y_test,rf_pred))
```
